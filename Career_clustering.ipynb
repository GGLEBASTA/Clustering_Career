```python
X = data.iloc[:, 0:-1]
```

# Data Analyze

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in X.columns:
    plt.figure(figsize=(7,3))
    sns.histplot(X[col], color="mediumspringgreen")
    plt.show()
```

```python
plt.figure(figsize=(20,9))
sns.heatmap(X.corr(), cmap="coolwarm_r", vmin=-1, annot=True, fmt=".2f")
plt.show()
```

# Data Scaler

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()
cols = X.columns
X = ms.fit_transform(X)
X = pd.DataFrame(X, columns=[cols])
X.head()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN().fit(X)
labels = dbscan.labels_

pd.Series(dbscan.labels_).unique()
```

```python
data['Groups'] = labels
data.sort_values('Groups', inplace=True)
data.head()
```

# Analyze results

```python
colors = ["black", "crimson", "blue", "green", "lime"]
fig, ax = plt.subplots(ncols=5, nrows=2, sharey=True, figsize=(30, 11), dpi=300)

k1, k2 = 0, 0
for i in data.columns[:-2]:
    if (k1 <= 4):
        sns.scatterplot(x=data['Groups'], y=data[i], ax=ax[0,k1] ,hue=pd.Series(dbscan.labels_), palette=colors)
        ax[0,k1].set_title(f"{i} & Groups")
        k1 = k1 + 1
    elif (k1 > 4):
        sns.scatterplot(x=data['Groups'], y=data[i], ax=ax[1,k2] ,hue=pd.Series(dbscan.labels_), palette=colors)
        ax[1,k2].set_title(f"{i} & Groups")
        k2 = k2 + 1

plt.show()
```

# Clustering Groups

```python
for i in pd.Series(dbscan.labels_).unique():
    careers = data[data['Groups'] == i]['Career'].values
    print(f'\n- - - Group {i} with size {len(careers)}:\n')
    print(careers)
```

# KNN

```python
from sklearn.cluster import KMeans

clusters = 5
km = KMeans(n_clusters=clusters)
km.fit(X)

pd.Series(km.labels_).unique()
```

```python
data['Groups_2'] = km.labels_
data.sort_values('Groups_2', inplace=True)
data.head()
```

```python
for i in pd.Series(km.labels_).unique():
    careers = data[data['Groups_2'] == i]['Career'].values
    print(f'\n- - - Group {i} with size {len(careers)}:\n')
    print(careers)
```
